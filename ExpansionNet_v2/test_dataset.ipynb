```python
from data.coco_dataset import CocoDatasetKarpathy
images_path = './ktvic_dataset/train_images/'
captions_path = './ktvic_dataset/'


coco_dataset = CocoDatasetKarpathy(
        images_path= images_path,
        coco_annotations_path=captions_path + "train_data.json",
        preproc_images_hdf5_filepath=None,
        precalc_features_hdf5_filepath=None ,
        limited_num_train_images=3000,
        limited_num_val_images=768)
```

```python
import pickle
with open('./demo_material/ktvic.pickle', 'wb') as f:
        coco_tokens = {'word2idx_dict':  coco_dataset.caption_word2idx_dict ,  'idx2word_list': coco_dataset.caption_idx2word_list, 'sos_str':  coco_dataset.get_sos_token_str()  , 'eos_str':   coco_dataset.get_eos_token_str()} 
        pickle.dump(coco_tokens, f)
```
